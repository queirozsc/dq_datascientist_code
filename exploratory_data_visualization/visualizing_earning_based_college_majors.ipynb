# Visualizing Earnings Based on College Majors

We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data.

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:
+ `[] Rank` - Rank by median earnings (the dataset is ordered by this column)
+ `[] Major_code` - Major code
+ `[] Major_category` - Category of major
+ `[] Total`- Total number of people with major
+ `[] Sample_size` - Sample size (unweighted) of full-time
+ `[] Men` - Male graduates
+ `[] Women`- Female graduates
+ `[] ShareWomen` - Women as share of total
+ `[] Employed` - Number employed
+ `[] Median` - Median salary of full-time, year-round workers
+ `[] Low_wage_jobs` - Number in low-wage service jobs
+ `[] Full_time` - Number employed 35 hours or more
+ `[] Part_time` - Number employed less than 35 hours

```python
# setting up the environment
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# reading the dataset
recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.iloc[0]
```

```python
# first rows
recent_grads.head()
```

```python
# last rows
recent_grads.tail()
```

```python
# summary statistics for all of the numeric columns
recent_grads.describe()
```

```python
# looking up the number of rows in dataset
raw_data_count = recent_grads.shape[0]
raw_data_count
```

```python
# dropping rows containing missing values
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
cleaned_data_count
```

## Exploring some relations

```python
relations = [("Sample_size", "Median"),
             ("Sample_size", "Unemployment_rate"),
             ("Full_time", "Median"),
             ("ShareWomen", "Unemployment_rate"),
             ("Men", "Median"),
             ("Women", "Median")]
for r in relations:
    recent_grads.plot(x=r[0], y=r[1], kind='scatter', 
                      title="{0} x {1}".format(r[0], r[1]))
```

## Do students in more popular majors make more money?

```python
recent_grads.plot(x="Total"
                  , y="Median"
                  , kind="scatter"
                 , title="Total number of people x Median salary")
```

## Do students that majored in subjects that where majority female make more money?

```python
recent_grads.plot(x="ShareWomen"
                  , y="Median"
                  , kind="scatter"
                 , title="Share of women x Median salary")
```

## Is there any link between the number of full-time employees and median salary?

```python
recent_grads.plot(x="Full_time"
                  , y="Median"
                  , kind="scatter"
                 , title="Full-time employees x Median salary")
```

### Grouped bar plot to compare the number of men with the number of women in each category of majors

```python
import numpy as np
(recent_grads
    .pivot_table(index="Major_category", values=["Men", "Women"], aggfunc=np.sum)
    .plot.bar())
```

### Box plot to explore the distributions of median salaries and unemployment rate

```python
color = dict(boxes="DarkGreen", whiskers="DarkOrange",
            medians="DarkBlue", caps="Gray")
plt.figure(figsize=(7,7))
(recent_grads["Median"]
    .plot.box(vert=False, color=color, sym="r+"))
```

```python
color = dict(boxes="DarkGreen", whiskers="DarkOrange",
            medians="DarkBlue", caps="Gray")
plt.figure(figsize=(7,7))
(recent_grads["Unemployment_rate"]
    .plot.box(vert=False, color=color, sym="r+"))
```

### Hexagonal bin plot to visualize the columns that had dense scatter plots

```python
relations = [("Sample_size", "Median"),
             ("Sample_size", "Unemployment_rate"),
             ("Full_time", "Median"),
             ("ShareWomen", "Unemployment_rate"),
             ("Men", "Median"),
             ("Women", "Median")]
for r in relations:
    recent_grads.plot.hexbin(x=r[0], y=r[1], gridsize=25, 
                      title="{0} x {1}".format(r[0], r[1]))
```
